```python
x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
#Av array ka bag of words converted to text to numbers  of the text column
cv = CountVectorizer() #here countvectorizer converted to objects
X = cv.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
print(X_train)
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
